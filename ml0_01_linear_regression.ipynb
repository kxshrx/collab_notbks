```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/COLLAB/01 machine learning/datasets/canada_per_capita_income.csv')
df
```

```python
df.describe()
corr_coeff = df['year'].corr(df['per capita income (US$)'])
corr_coeff
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.3,random_state = 0)
print("xtrain",xtrain,"\nxtest",xtest,"\nytrain",ytrain,"\nytest",ytest)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
y_pred = regressor.predict(xtest)
y_pred
```

```python
regressor.predict([[2000]])
```

```python
plt.scatter(xtrain,ytrain,color='red')
plt.plot(xtrain,regressor.predict(xtrain),color='green')
```

```python
plt.scatter(xtest,ytest,color='black')
plt.plot(xtest,regressor.predict(xtest),color='blue')
```

```python
plt.scatter(xtest, ytest, color='black')
plt.plot(xtest, regressor.predict(xtest), color='blue')
plt.scatter(xtrain, ytrain, color='red')
plt.plot(xtrain, regressor.predict(xtrain), color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot for test data
plt.scatter(xtest, ytest, color='black', label='Test Data')

# Regression line for test data
plt.plot(xtest, regressor.predict(xtest), color='blue', linestyle='--', label='Test Prediction')

# Scatter plot for training data
plt.scatter(xtrain, ytrain, color='red', label='Training Data')

# Regression line for training data
plt.plot(xtrain, regressor.predict(xtrain), color='green', linestyle='-', label='Training Prediction')

# Add titles and labels
plt.title('Linear Regression Model')
plt.xlabel('Feature')
plt.ylabel('Target')

# Add a legend
plt.legend()

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'year' is your x variable and 'xtest' and 'xtrain' are the year data for test and train sets
# 'ytest' are the actual target values for the test set
# 'ytrain' are the actual target values for the training set
# 'regressor' is your trained regression model

# Create the plot
plt.figure(figsize=(12, 8))

# Scatter plot for actual training data
plt.scatter(xtrain, ytrain, color='red', label='Training Data Actual')

# Plot for predicted training data
plt.plot(xtrain, regressor.predict(xtrain), color='green', linestyle='-', label='Training Data Predicted')

# Scatter plot for actual test data
plt.scatter(xtest, ytest, color='black', label='Test Data Actual')

# Plot for predicted test data
plt.plot(xtest, regressor.predict(xtest), color='blue', linestyle='--', label='Test Data Predicted')

# Add titles and labels
plt.title('Actual vs Predicted Values Over Years')
plt.xlabel('Year')
plt.ylabel('Target Variable')

# Add a legend
plt.legend()

# Add a grid
plt.grid(True)

# Show the plot
plt.show()

```
